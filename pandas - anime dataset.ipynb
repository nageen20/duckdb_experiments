```python
import pandas as pd
import time
import numpy as np
```

```python
%%time
# Reading the user details anime dataset
user_details = pd.read_csv('data/users-details-2023.csv')
```

```python
%%time
user_details
```

```python
%%time
# Reading the user ratings and anime details datasets
user_ratings = pd.read_csv('data/users-score-2023.csv')
anime_details = pd.read_csv('data/anime-dataset-2023.csv')
```

```python
%%time
user_ratings
```

```python
%%time
anime_details
```

```python
%%time
# Join anime_details, user_ratings and user details datasets and get the required columns
anime_ratings = anime_details[['anime_id','Name','Genres','Type','Episodes','Duration','Aired']].\
                    merge(user_ratings[['user_id','anime_id','rating']], left_on='anime_id', right_on='anime_id', how='left').\
                    merge(user_details[['Mal ID','Username','Gender','Birthday','Location','Mean Score','Watching',
                                       'Completed','On Hold','Dropped','Episodes Watched']], left_on='user_id', right_on='Mal ID', how='left')
anime_ratings['genres_list'] = anime_ratings['Genres'].str.split(',')
```

```python
%%time
# Re-order the columns
anime_ratings = anime_ratings[['anime_id','Name','Genres','genres_list','Type','Episodes','Duration','Aired','Mal ID','Username','Gender','Birthday',
                            'Location','Mean Score','Watching','Completed','On Hold','Dropped','Episodes Watched','rating']]
```

```python
%%time
# Rename the columns
anime_ratings.rename(columns = {'Name': 'anime_name', 'Genres': 'genres', 'Type': 'type', 'Episodes': 'episodes', 'Duration':'episode_duration',
                                'Aired': 'aired_dates', 'Mal ID': 'user_id', 'Username':'user_name', 'Gender': 'user_gender',
                                'Birthday': 'user_birthday', 'Location': 'user_location', 'Mean Score': 'mean_score', 'Watching':'animes_watching_user',
                                'Completed':'animes_completed_user', 'On Hold':'animes_onHold_user', 'Dropped':'animes_dropped_user',
                                'Episodes Watched':'total_animes_user_watched', 'rating':'user_rating'
                    }, inplace=True)
```

```python
%%time
anime_ratings
```

```python
%%time
# Total rows in the table
start_time = time.time()
print(anime_ratings.count())
end_time = time.time()
print(f'Total time taken: {end_time - start_time}')
```

```python
%%time
# Get the distinct anime in the file and their average ratings sort by rating descending
start_time = time.time()
avg_ratings = anime_ratings.groupby(['anime_name']).agg({'user_rating':np.mean, 'user_id': lambda x: x.nunique()}).sort_values('user_rating',ascending=False)

display(avg_ratings)
end_time = time.time()
print(f'Total time taken: {end_time - start_time}')
```

We see that in the above output, there are a lot of animes where very few users gave ratings. The top animes we see in the above output are only rated by just one user. Let us filter to anime that are rated by atleast 1000 users so we can get a better list.

```python
%%time
# Get the distinct anime in the file and their average ratings 
# sort by rating descending for anime that have more than 1000 user ratings
start_time = time.time()
avg_ratings = anime_ratings.groupby(['anime_name']).\
                            filter(lambda x: x['user_id'].nunique() > 1000).\
                            groupby('anime_name')['user_rating'].mean().sort_values(ascending=False)

display(avg_ratings)
end_time = time.time()
print(f'Total time taken: {end_time - start_time}')
```

```python
%%time
# Get the distinct anime  genres in the file and their average ratings
start_time = time.time()
avg_ratings_genre = anime_ratings.explode('genres_list').\
                    groupby(['genres_list']).\
                    agg({'anime_id': lambda x: x.nunique(),
                                                                                      'user_rating': np.mean,
                                                                                      'user_id': lambda x: x.nunique()}).\
                                                                                    sort_values('user_rating',ascending=False)

display(avg_ratings_genre)
end_time = time.time()
print(f'Total time taken: {end_time - start_time}')
```
